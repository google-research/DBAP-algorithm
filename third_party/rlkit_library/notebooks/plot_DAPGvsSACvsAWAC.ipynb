```python
import matplotlib.pyplot as plt
%matplotlib inline
import csv
import pickle
import numpy as np
truncate_val = 250
DAPG_result_paths = ['/usr/local/google/home/abhishekunique/sim_franka/mjrl/examples/2element_DAPG/iterations/all_stats.pickle',
                    '/usr/local/google/home/abhishekunique/sim_franka/mjrl/examples/2element_DAPG1/iterations/all_stats.pickle',
                    '/usr/local/google/home/abhishekunique/sim_franka/mjrl/examples/2element_DAPG2/iterations/all_stats.pickle']

overall_vals_DAPG = [[] for _ in range(len(DAPG_result_paths))]
for j, path in enumerate(DAPG_result_paths):
    result = pickle.load(open(path, 'rb'))
    for iter_stat in result:
        overall_vals_DAPG[j].append(iter_stat['overall_eval'])

for j, ovd in enumerate(overall_vals_DAPG):
    ovd = np.asarray(ovd[:truncate_val])
    overall_vals_DAPG[j] = ovd

overall_vals_DAPG = np.array(overall_vals_DAPG)
```

```python
import csv

SAC_paths = ['/usr/local/google/home/abhishekunique/sim_franka/corl_data/20210615-16h59m-sac-withresets-2elements-cabinetslider/test-0/test-0/06-16-dev-example-sac-script/06-16-dev-example-sac-script_2021_06_16_00_04_29_0000--s-4070/progress.csv', 
              '/usr/local/google/home/abhishekunique/sim_franka/corl_data/20210615-16h59m-sac-withresets-2elements-cabinetslider/test-1/test-1/06-16-dev-example-sac-script/06-16-dev-example-sac-script_2021_06_16_00_03_08_0000--s-70874/progress.csv',
              '/usr/local/google/home/abhishekunique/sim_franka/corl_data/20210615-16h59m-sac-withresets-2elements-cabinetslider/test-2/test-2/06-16-dev-example-sac-script/06-16-dev-example-sac-script_2021_06_16_00_03_36_0000--s-75743/progress.csv']
overall_vals_SAC = [[] for _ in range(len(SAC_paths))]
for j, path in enumerate(SAC_paths):
    with open(path, mode='r') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        line_count = 0
        for row in csv_reader:
            if line_count == 0:
                print(f'Column names are {", ".join(row)}')
                line_count += 1
            overall_vals_SAC[j].append(float(row['eval_exhaustivess/overall_eval']))
            line_count += 1
        print(f'Processed {line_count} lines.')

for j, ovs in enumerate(overall_vals_SAC):
    ovs = np.asarray(ovs[:truncate_val])
    overall_vals_SAC[j] = ovs

overall_vals_SAC = np.array(overall_vals_SAC)
```

```python
import csv

AWAC_paths = ['/usr/local/google/home/abhishekunique/sim_franka/corl_data/20210614-20h28m-playdata-2elements-rerun-june14-cabinetslider/test-8/test-8/06-15-dev-example-awac-script/06-15-dev-example-awac-script_2021_06_15_03_45_16_0000--s-83532/progress.csv', 
              '/usr/local/google/home/abhishekunique/sim_franka/corl_data/20210614-20h28m-playdata-2elements-rerun-june14-cabinetslider/test-9/test-9/06-15-dev-example-awac-script/06-15-dev-example-awac-script_2021_06_15_03_44_40_0000--s-58009/progress.csv',
              '/usr/local/google/home/abhishekunique/sim_franka/corl_data/20210614-20h28m-playdata-2elements-rerun-june14-cabinetslider/test-11/test-11/06-15-dev-example-awac-script/06-15-dev-example-awac-script_2021_06_15_03_44_05_0000--s-14880/progress.csv']
overall_vals_AWAC = [[] for _ in range(len(AWAC_paths))]
for j, path in enumerate(AWAC_paths):
    with open(path, mode='r') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        line_count = 0
        for row in csv_reader:
            if line_count == 0:
                print(f'Column names are {", ".join(row)}')
                line_count += 1
            overall_vals_AWAC[j].append(float(row['eval_exhaustivess/overall_eval']))
            line_count += 1
        print(f'Processed {line_count} lines.')
    
for j, ova in enumerate(overall_vals_AWAC):
    ova = np.asarray(ova[:truncate_val])
    overall_vals_AWAC[j] = ova

overall_vals_AWAC = np.array(overall_vals_AWAC)
```

```python
truncate_val = 250
import seaborn as sns
sns.set_theme()

# for dapg_line in overall_vals_DAPG:
plt.plot(np.mean(np.array(overall_vals_DAPG), axis=0), color='r', marker='o', label='DAPG')
plt.fill_between(range(truncate_val), 
                 np.mean(np.array(overall_vals_DAPG), axis=0) - np.std(np.array(overall_vals_DAPG), axis=0),  
                 np.mean(np.array(overall_vals_DAPG), axis=0) + np.std(np.array(overall_vals_DAPG), axis=0),  
                 color='r', alpha=0.5)

# for sac_line in overall_vals_SAC:
plt.plot(np.mean(np.array(overall_vals_SAC), axis=0), color='b', marker='x',  label='SAC')
plt.fill_between(range(truncate_val), 
             np.mean(np.array(overall_vals_SAC), axis=0) - np.std(np.array(overall_vals_SAC), axis=0),  
             np.mean(np.array(overall_vals_SAC), axis=0) + np.std(np.array(overall_vals_SAC), axis=0),  
             color='b', alpha=0.5)

# for awac_line in overall_vals_AWAC:
plt.plot(np.mean(np.array(overall_vals_AWAC), axis=0), color='g', marker='+', label='AWAC')
plt.fill_between(range(truncate_val), 
             np.mean(np.array(overall_vals_AWAC), axis=0) - np.std(np.array(overall_vals_AWAC), axis=0),  
             np.mean(np.array(overall_vals_AWAC), axis=0) + np.std(np.array(overall_vals_AWAC), axis=0),  
             color='g', alpha=0.5)
plt.legend()
plt.savefig("singletask_comparisons.png")
```
